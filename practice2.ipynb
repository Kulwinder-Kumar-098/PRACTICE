```python
import numpy as np

X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Add intercept
X = np.c_[np.ones(len(X)), X]
```

## β^​=(XTX)^-1XTy

```python
beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y
beta_hat
```

```python
y_pred = X @ beta_hat
```

```python
residuals = y - y_pred
n, p = X.shape

sigma_squared = (residuals @ residuals) / (n - p)
sigma = np.sqrt(sigma_squared)
```
