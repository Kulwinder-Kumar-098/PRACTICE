```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np

# Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([0, 0, 0, 1, 1])

# Initialize parameters
w = 1.0
b = -3.0
lr = 0.1
epochs = 1000

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Training loop
for _ in range(epochs):
    z = w * X + b
    p = sigmoid(z)

    # Gradients
    dw = np.mean((p - y) * X)
    db = np.mean(p - y)

    # Update
    w -= lr * dw
    b -= lr * db

print("Trained weight:", w)
print("Trained bias:", b)
```

```python
def predict(x):
    z = w * x + b
    p = sigmoid(z)
    return 1 if p >= 0.5 else 0

print(predict(2))  # Expect 0
print(predict(5))  # Expect 1
```

### Model preparation and evaluation
We will train a logistic regression classifier on the breast cancer dataset, standardize features, and compare performance on training and test data.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer

X = df.drop(columns=['diagnosis'])
y = df['diagnosis']

# Report missing values before training
missing_counts = X.isnull().sum()
print("Missing values per column:")
print(missing_counts[missing_counts > 0])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

model = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler(),
    LogisticRegression(max_iter=1000, random_state=42)
)
model.fit(X_train, y_train)

print("Training accuracy: {:.3f}".format(model.score(X_train, y_train)))
print("Test accuracy: {:.3f}".format(model.score(X_test, y_test)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

predictions = model.predict(X_test)

print("Classification Report:\n")
print(classification_report(y_test, predictions, target_names=['Benign', 'Malignant']))

cm = confusion_matrix(y_test, predictions)
print("Confusion matrix:\n", cm)

plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

### Conclusion
